```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits= pd.read_csv("tmdb_5000_credits.csv")
movies.sample(5)
```

```python
movies=movies.merge(credits,on='title') #merging credits to movies on the basis of titles
```

```python
movies.shape
#we'll be keeping
movies=movies[['movie_id','title','overview','keywords','genres','cast','crew']]
movies
```

```python
movies.isnull().sum()
```

```python
# as 3 missing values is not a big number we will drop them
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
#bringing all the columns into a correct format
movies.iloc[0].genres
```

```python
import ast  #String type values will be converted in LIST. 

def convert(obj):
    namelist=[]
    for i in ast.literal_eval(obj):
        namelist.append(i['name'])  #the name key of the genre column will be appended
    return namelist
movies['genres']=movies['genres'].apply(convert)
movies
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies
```

```python
import ast  #String type values will be converted in LIST. 

def convert2(obj):
    namelist=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            namelist.append(i['name']) # only the top 3 actors
            counter+=1
        else:
            break
       
    return namelist
```

```python
movies['cast']=movies['cast'].apply(convert2)
movies
```

```python

def fetch_director(obj):
    namelist=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
             namelist.append(i['name'])  #Only directors will be appended
             break   
    
            
       
    return namelist
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies
```

```python
#The split() method splits a string into a list. 
#the 'overview' column is a string so we will convert it in LIST usinf split()

movies['overview']=movies['overview'].apply(lambda x:x.split())
movies
```

```python
#as we have converted all the string columns into LIST .
#concatinate all the lists together and make one column
```

```python
#using transformer we will remove the space between the words (Sam Mendes --> SamMendes) and it one word
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies
```

```python
new_df=movies[['movie_id' ,'title', 'tags']]
```

```python
new_df['tags']= new_df['tags'].apply(lambda x:(" ").join(x))  #list to string
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
 #tags -> vectors. of a user chooses a movie(vectors) then all the closest vectors will be similar and will be recommended
```

```python
#Removing the stop words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors  #every movie is now in a vector form
```

```python
cv.get_feature_names()  #now here some words are --> 'american' & 'americans' , 'activity' & 'activities'
#we will bring all the similar words to the root word using stemming
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stemming(text):
    l=[]
    for i in text.split(): #converting string to List for stemming
        stem=ps.stem(i)
        l.append(stem)
        
    return " ".join(l)  # converting List to string back
new_df['tags'] = new_df['tags'].apply(stemming)
new_df['tags'][0]
```

```python
cv.get_feature_names()
```

```python
#as 5000 vectors are plotted on the dimentionality space. to find the distance between each vector we will use "COSINE SIMILARITY'
#instead of "EUCLIDIAN DISTANCE". eauclidian distance fails in high dimentional data
#larger the angle more distinct the movies are
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)  # 1--> exact similar  0 --> distinct
```

```python
#sorted(similarity[0],reverse=True)  #similarity of first movie with all 4806 movies
```

```python
#Enumerate() method adds a counter to an iterable and returns it in a form of enumerating object. 
#as we are sorting the similarities we are losing the indexes so we are using the enumerate() function

sorted(list(enumerate(similarity[0])),reverse= True, key=lambda x:x[1])[1:6]

#key=lambda x:x[1] sorted the values and not the indexes
```

```python
def recommed(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse= True, key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
       
```

```python
recommed('Batman Begins')
```

```python
recommed('Gandhi')
```
